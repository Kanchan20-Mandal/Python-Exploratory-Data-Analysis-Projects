```python
# create a function to remove outliers
def remove_outliers(a, data, mini, maxi):
    upper_array = np.where(data[a] <= mini)[0]
    lower_array = np.where(data[a] >= maxi)[0]

    # Convert NumPy indices to actual DataFrame index labels
    upper_labels = data.index[upper_array] # upper_labels = [10,14,18]
    lower_labels = data.index[lower_array]

    print(f"Old shape:", data.shape)
    data.drop(index=upper_labels, inplace=True, errors='ignore')
    data.drop(index=lower_labels, inplace=True, errors='ignore')
    print(f"New shape:", data.shape)
```

```python
# Create a function that identify outliers and removes them. Data represents dataframe and a represents column name
def identify_outliers(data,a): 
    # calculate quartile 1 which represents 25th percentile data
    q1 = np.percentile(data[a],25,method="midpoint")
    # calculate quartile 2 which represents 50th percentile data
    q2 = np.percentile(data[a],50,method="midpoint")
    # calculate quartile 3 which represents 75th percentile data
    q3 = np.percentile(data[a],75,method="midpoint")

    # inter quartile range 
    iqr = q3-q1

    # caluclate minimum and maximum values for this data
    mini = q1 - 1.5*iqr
    maxi = q3 + 1.5*iqr

    # print data 
    print(a)

    # print the iqr, mini and maxi ranges for this data
    print("IQR","\t","Minimum","\t","Maximum")
    print(iqr,'\t',mini,'\t',maxi)

    remove_outliers(a,data,mini,maxi)

    print("**********************************")
    
```

```python
for i in con:
    identify_outliers(lp,i)
```
